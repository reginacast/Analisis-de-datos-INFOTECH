# **EJEMPLO INTEGRADOR 1**

Se aplican los siguientes conceptos básicos de Python:
- Tipos de datos
- Operadores arimeticos
- Operador de asignacion
- Operadores de relaciones y logicos
- if / elif / else
- For
- While
- While + else
- iterables e iteradores
- For anidado
- While anidado



```python
##############################################################################################################
#1. Tipos de datos
nombre_tienda = "Tienda de Regina" #string
abierta = True             #Bool
meta_diaria = 100.0        #Float
dias = ("Lunes", "Martes", "Miércoles", "Jueves", "Viernes") #Tupla

ventas_semana = [
    [40, 30, 20],
    [ 100, 300, 200],
    [250, 180, 200],
    [0, 0, 0],
    [800, 200, 320]
] #list de listas

info = {"Nombre de la Tienda": nombre_tienda, "Objetivo diario de venta": meta_diaria} #diccionario
categorias = {"bebidas", "snacks", "fruta"} #set

print("\n -----------Información básica----------")

print("Información: ", info)
print("Categorías en venta: ", categorias)


##############################################################################################################
#2. Operadores aritmericos

# Inicializamos el acumulador para la suma de toda la semana
total_semana = 0
# Creamos una lista vacía para guardar el resultado final de cada día
totales_por_dia = []

# Iniciamos un ciclo que recorrerá los índices (0 a 4) de la lista ventas_semana
for dia in range(len(ventas_semana)):
  # Sumamos los tres elementos de la sub-lista actual.
  # [dia] elige la sub-lista (el día) y [0],[1],[2] las ventas individuales de ese día.
  total_dia = ventas_semana[dia][0] + ventas_semana[dia][1] + ventas_semana[dia][2]

  # Acumulamos el total del día actual al gran total de la semana (usando el operador +=)
  total_semana += total_dia

  # Guarda el resultado del día actual al final de la lista para no perderlo.
  totales_por_dia.append(total_dia)

# Mostramos los resultados finales procesados
print("Venta total por día: ", totales_por_dia)
print("Venta total de la semana: ", total_semana)


##############################################################################################################
# 3. If / Elif / Else (Evaluación de metas)

print("\n -----------Evaluacion de metas----------")

# Iniciamos un ciclo que usa 'i' como índice para recorrer los días (0 a 4)
for i in range (len(totales_por_dia)):

  # CASO 1: Si el total del día 'i' es exactamente igual a 0
  if totales_por_dia[i] == 0:
    # dias[i] nos da el nombre (ej. "jueves") y le pega el texto "Sin ventas"
    print(dias[i], "SIN VENTAS")

  # CASO 2: Si no fue 0, pero el total es mayor o igual (>=) a la meta (100.0)
  elif totales_por_dia[i] >= meta_diaria:
    print(dias[i], "objetivo diario de venta CUMPLIDO")

  # CASO 3: Si no fue ninguna de las anteriores (entonces el total es mayor a 0 pero menor a 100)
  else:
    print(dias[i], "objetivo diario de venta NO CUMPLIDO")


##############################################################################################################
#4. For anidado


# Definimos el orden de las categorías (debe coincidir con el orden de tus listas de ventas)
categoria = ["bebidas", "snacks", "fruta"]

max_venta = 0
categoria_ganadora = ""

# El primer ciclo recorre cada día de la semana
for fila in ventas_semana:

    # El segundo ciclo usa 'range' para saber en qué columna (0, 1 o 2) estamos parados
    for i in range(len(fila)):
        venta_actual = fila[i] # Extraemos el número de la posición 'i'

        # Si este número es el más grande que hemos visto hasta ahora...
        if venta_actual > max_venta:
            max_venta = venta_actual          # Guardamos el número
            categoria_ganadora = categoria[i] # Guardamos el nombre de la categoría usando el mismo índice 'i'

# Ahora el print nos dará toda la información
print(f"La categoría {categoria_ganadora} tuvo el máximo número de ventas con un total de {max_venta}")


##############################################################################################################
##############################################################################################################
# 5. While (Bucle basado en una condición)

print("\n ---Detección de día que no hubo ventas---")

# Creamos manualmente una variable que servirá como contador (puntero)
indice = 0

# La condición: Mientras el índice sea menor al tamaño de la lista (5)...
while indice < len(totales_por_dia):

    # Imprimimos el nombre del día y su total usando el índice actual
    print("Venta total del día",dias[indice],":",totales_por_dia[indice])

    # ¡VITAL! Incrementamos el índice en 1. Sin esto, el índice siempre sería 0 y el programa nunca terminaría.
    indice += 1


##############################################################################################################
#6. While + else

indice = 0 # Reiniciamos el contador a 0

while indice < len(totales_por_dia):
    # Revisamos si el total de ese día es cero
    if totales_por_dia[indice] == 0:
        print("Día sin ventas detectado: ", dias[indice])
        break  # ¡ALTO! Ya encontramos lo que buscábamos, detén el bucle por completo.

    indice += 1 # Avanzamos al siguiente día si no fue cero
else:
    # Este 'else' pertenece al 'while'.
    # Solo se ejecuta si el bucle terminó NORMALMENTE (o sea, si NUNCA se activó el break).
    print("Todos los días tuvieron ventas")

##############################################################################################################
#7. Iterable e Iterador (Recorrido manual)

print("\n -----------Iterador de días----------")

# Convertimos la tupla 'dias' en un objeto iterador llamado 'it'
# El iterador tiene el control de en qué posición de la lista estamos.
it = iter(dias)

# Cada vez que llamamos a next(), el iterador nos entrega el siguiente valor
# y se queda "esperando" en esa posición para la próxima llamada.
print(next(it)) # Muestra: lunes
print(next(it)) # Muestra: martes
print(next(it)) # Muestra: miercoles
print(next(it)) # Muestra: jueves
print(next(it)) # Muestra: viernes

# Nota: Si llamaras a next(it) una sexta vez, el programa daría un error (StopIteration)
# porque ya no quedan más elementos en la tupla.
```
