```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

**游늷 An치lisis Descriptivo: 쯈u칠 pas칩?**

El objetivo es resumir y describir los datos tal como son, sin intentar adivinar nada m치s.

```python
# Objetivo: Resumir qu칠 pas칩 en el pasado.
def ejemplo_descriptiva():
    # Paso A: # Fijar una "semilla" para que los n칰meros aleatorios sean los mismos siempre
    np.random.seed(0)

    # Paso B: Crear una tabla (DataFrame) simulando datos crudos
    # Generamos 100,000 registros de horas (del 0 al 23)
    datos = pd.DataFrame({
        "hora": np.random.randint(0, 24, 100_000),
    })

    # Paso C: Transformar datos crudos en informaci칩n 칰til
    # 1. value_counts(): Cuenta cu치ntos coches pasaron por cada hora (ej. Hora 5: 4000 veces)
    # 2. sort_index(): Ordena el resultado del 0 al 23 para que tenga sentido cronol칩gico
    conteo_por_hora = datos["hora"].value_counts().sort_index()

    print("\n=== INTERPRETACI칍N DESCRIPTIVA ===")
    print("Veh칤culos detectados por hora:")
    print(conteo_por_hora) # Esto imprime una tabla de frecuencias
```

```python
ejemplo_descriptiva()
```

**游늷An치lisis Inferencial: De una muestra a la poblaci칩n**

Aqu칤 usamos una peque침a parte de los datos (muestra) para sacar conclusiones sobre un grupo mucho m치s grande.

```python
# Objetivo: Usar una muestra peque침a para adivinar el total de una poblaci칩n.
def ejemplo_inferencial():
    np.random.seed(1) # Usamos el n칰mero 1 para que esta simulaci칩n use una "lista de azar"

    # Paso A: Tomamos una "muestra" de 5,000 personas
    # Simulamos que el promedio de uso es 80 minutos con una desviaci칩n de 20
    muestra = pd.DataFrame({
        "tiempo_uso_min": np.random.normal(80, 20, 5000)
    })

    # Paso B: Calculamos el promedio de estos 5,000 para "inferir" c칩mo se comporta el resto
    media_muestra = muestra["tiempo_uso_min"].mean()

    print("\n=== INTERPRETACI칍N INFERENCIAL ===")
    # Concluimos que, aunque no medimos a todos, el promedio poblacional debe rondar este n칰mero
    print(f"Tiempo promedio estimado en toda la poblaci칩n: {media_muestra:.2f} min/d칤a")
```

```python
ejemplo_inferencial()
```

**游늷 An치lisis Comparativo: 쮿ay diferencias entre grupos?**

Comparamos dos o m치s grupos para ver si su comportamiento var칤a significativamente.

```python
# Objetivo: Ver si hay diferencias reales entre dos situaciones.
def ejemplo_comparativa():
    np.random.seed(2)

    # Paso A: Creamos dos grupos (D칤a Normal vs Black Friday)
    # Generamos 100k filas de "normal" y 100k de "black_friday"
    trafico = pd.DataFrame({
        "evento": ["normal"] * 100_000 + ["black_friday"] * 100_000,
        # np.random.poisson genera n칰meros basados en un promedio (300 y 900)
        "peticiones": np.concatenate([
            np.random.poisson(300, 100_000),   # Promedio de 300 peticiones
            np.random.poisson(900, 100_000),   # Promedio de 900 peticiones
        ])
    })

    # Paso B: Agrupamos por tipo de evento y sacamos el promedio de cada uno para comparar
    comparacion = trafico.groupby("evento")["peticiones"].mean()

    print("\n=== INTERPRETACI칍N COMPARATIVA ===")
    print("Peticiones promedio por evento:")
    print(comparacion) # Aqu칤 veremos claramente que Black Friday es 3 veces m치s alto
```

```python
ejemplo_comparativa()
```

**游늷 An치lisis Causal: 쯇or qu칠 pas칩? (Causa y Efecto)**

Busca identificar si un cambio espec칤fico (una versi칩n de algoritmo) es el responsable de un resultado (clicks).

```python
 # Objetivo: Probar si un cambio espec칤fico caus칩 un resultado diferente.
def ejemplo_causal():
    np.random.seed(3) # Para que los "clics" aleatorios sean iguales cada vez que corras el c칩digo
    n = 200_000       # Definimos un tama침o de muestra grande para que el resultado sea confiable

    # Paso A: Dividimos a los usuarios en dos grupos (v1 y v2 del algoritmo)
    # Creamos una lista donde los primeros 100k son "v1" y los otros 100k son "v2"
    version = np.array(["v1"] * (n // 2) + ["v2"] * (n // 2))

    # Paso B: Simulamos que v2 genera m치s clicks (promedio 4) que v1 (promedio 2)
    # Generamos clics aleatorios para cada grupo usando una distribuci칩n de Poisson
    # (v1 suele dar 2 clics, v2 suele dar 4 clics)
    clicks_v1 = np.random.poisson(2, n // 2)
    clicks_v2 = np.random.poisson(4, n // 2)

    # Paso C: Unimos todo en una tabla para analizar
    datos_algoritmo = pd.DataFrame({
        "version_algoritmo": version,
        "clicks": np.concatenate([clicks_v1, clicks_v2]) # Pegamos las dos listas de clics
    })
    # Paso D: Agrupamos por versi칩n para comprobar si el cambio de versi칩n "caus칩" la mejora en clics
    resultado = datos_algoritmo.groupby("version_algoritmo")["clicks"].mean()

    print("\n=== INTERPRETACI칍N CAUSAL (SIMULADA) ===")
    print("Clicks promedio por versi칩n de algoritmo:")
    print(resultado)
    print("Interpretaci칩n: la versi칩n v2 parece incrementar los clics promedio.")
```

```python
ejemplo_causal()
```

**游늷 An치lisis Predictivo: 쯈u칠 pasar치?**

Usa datos hist칩ricos para entrenar un modelo que pueda "adivinar" el futuro o valores desconocidos.

```python
# Objetivo: Entrenar un modelo para que adivine el futuro.
def ejemplo_predictiva():
    np.random.seed(4)

    # Paso A: Generar datos hist칩ricos (Relaci칩n: a m치s calor, m치s consumo de energ칤a)
    temperatura = np.random.uniform(5, 35, 1000) # 1000 d칤as con temperaturas de 5 a 35 grados
    # Creamos una demanda de energ칤a que DEPENDE de la temperatura:
    # Demanda = Base(200) + (Temperatura * 8) + un poco de ruido/error aleatorio
    demanda = 200 + temperatura * 8 + np.random.normal(0, 10, 1000)

    # Paso B: Formatear los datos para el modelo (X es la entrada, y es el resultado)
    # Scikit-learn necesita que la 'X' (causa) sea una matriz (columna), por eso usamos .reshape
    X = temperatura.reshape(-1, 1)
    y = demanda # La 'y' es lo que queremos predecir (efecto)

    # Paso C: Crear y "entrenar" el modelo (aprende la relaci칩n temperatura/demanda)
    modelo = LinearRegression() # Elegimos el algoritmo (Regresi칩n Lineal: dibuja una l칤nea recta)
    modelo.fit(X, y) # El modelo "estudia" los datos para encontrar la f칩rmula: y = mx + b

    # Paso D: Usar el modelo para predecir algo que no ha pasado
    # Queremos saber: "Si ma침ana hace 30춿C, 쯖u치nta luz se va a gastar?"
    temp_objetivo = 30
    # Usamos el modelo entrenado para predecir ese valor espec칤fico
    demanda_predicha = modelo.predict([[temp_objetivo]])[0]

    print("\n=== INTERPRETACI칍N PREDICTIVA ===")
    print(f"Predicci칩n: Si hace {temp_objetivo}춿C, la demanda ser치 {demanda_predicha:.2f}")
```

```python
ejemplo_predictiva()
```

**游늷 An치lisis Prescriptivo: 쯈u칠 debemos hacer?**

Es el nivel m치s avanzado. No solo predice el futuro, sino que sugiere la mejor acci칩n para optimizar un resultado.

```python
# Objetivo: Evaluar opciones y recomendar la mejor acci칩n.
def ejemplo_prescriptiva():
    # Paso A: Datos de entrada sobre diferentes caminos (rutas)
    # Imaginemos 5 rutas diferentes para llegar al trabajo
    distancias = np.array([5, 12, 8, 3, 10])         # Kil칩metros de cada ruta
    tiempos_trafico = np.array([1.2, 1.5, 1.3, 1.0, 1.4]) # Multiplicador de tr치fico (1.5 = mucho tr치fico)

    # Paso B: Calcular el "costo" o tiempo final de cada opci칩n (Distancia x Tr치fico)
    # Calculamos una "puntuaci칩n" para cada ruta: Distancia por Tr치fico
    # El resultado es el tiempo estimado de llegada
    puntuacion = distancias * tiempos_trafico

    # Paso C: Seleccionar la opci칩n ganadora (la ruta con el valor m치s bajo)
    # .argmin() busca en la lista de puntuaciones y nos da la posici칩n (칤ndice) de la m치s peque침a
    ruta_optima = puntuacion.argmin()

    print("\n=== INTERPRETACI칍N PRESCRIPTIVA ===")

    # Recomendaci칩n final basada en el c치lculo
    print(f"Acci칩n recomendada: Toma la ruta {ruta_optima} porque es la m치s eficiente.")
```

```python
ejemplo_prescriptiva()
```
