## **Numpy:**




Proporciona las estructuras de datos básicas implementando matrices y arreglos multidimensionales. También proporciona las funcionalidades necesarias para crear y manipular estas estructuras de datos.

```python
import numpy as np
```

```python
valores_celsius = [20.1, 20.8, 21.9, 22.5, 22.7, 22.3, 21.8, 21.2, 20.9,20.1]
```

```python
#Ejemplo utilizando Numpy
C = np.array(valores_celsius) #Se convierte la lista valores_celsius en un arrelgo de una dimensión en Numpy
print(C)
print(type(C))

F = (C*9/5+31)
print(F) #Se realiza la conversión a grados Farenheit usando una operación en el arreglo y se imprime
print(type(F))
```

```python
#Ejemplo utilizando listas en vez de Numpy
valores_farenheit = [x*9/5+31 for x in valores_celsius] #Se itera sobre la lista para ir calculando sobre cada valor
print(valores_farenheit)
print(type(valores_farenheit))
```

## **Pandas:**

 Se utiliza para la manipulación y el análisis de datos. Asimismo, proporciona estructuras de datos y operaciones especiales para la manipulación de tablas numéricas y series temporales.

Las estructuras de datos más importantes de esta librería son *Series* y *DataFrame*

***Series***

- Es un objeto unidimensional
- Puede contener cualquier tipo de dato
- Contiene dos arreglos: uno que funciona como índice y el otro que contiene los datos

```python
import pandas as pd
```

```python
obj = pd.Series([4,7,-5,3])
obj

#El índice es la columna de lado izquierdo y los datos el del lado derecho
```

```python
#Para reemplazar un valor en nuestro arreglo
obj.replace(-5, 5, inplace=True) #X es el valor que queremos reemplazar, Y es el nuevo valor que queremos asignar (x,y)
                                 #Inplace modifica el arreglo directamente en vez de crear uno nuevo
obj
```

```python
# Acceso al índice y a los valores de la serie

print(obj.index)
print(obj.values)
```

```python
# Definición de objeto Serie con índices individuales

frutas = ['manzanas', 'naranjas', 'cerezas', 'peras'] #Creamos nuestro propio índice a través de una lista
cantidades = [20, 33, 52, 10] #Creamos nuestros propios datos a través de una lista
serie = pd.Series(cantidades, index=frutas) #Le indicamos cuál es nuestro índice y cuáles son los datos
serie
```

```python
#Si se suman dos series con el mismo índice, se obtiene una nueva serie con el mismo índice y los valores correspondientes se suman

serie = pd.Series(cantidades, index=frutas)
serie2 = pd.Series([17, 13, 31, 32], index=frutas) #le damos los datos que queremos que tenga la serie 2
print(serie + serie2) #sumamos ambas series

print("La suma de la primera serie es: ", sum(serie))
print("La suma de la segunda serie es: ", sum(serie2))
```

***DataFrame***

- Contiene una colección ordenada de columnas con nombre, cada una de las cuales puede tener un tipo d valor diferente.
- Tiene un índice de fila y uno de columna

```python
#Creación de un diccionario de datos

datos = {"estado":["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada", "Nebraska","Nebraska","Nebraska", "Texas","Texas","Texas"],
         "año":["2000", "2001", "2002", "2001", "2002", "2003","2000","2002","2004","2001","2002","2003"],
         "población (millones)":[1.5, 1.7, 3.6, 2.4, 2.9, 3.2, 1.3, 2.6, 4.3, 2.3, 4.5, 6.7]}

print(type(datos))
```

```python
#Para convertir un diccionario en un DataFrame

frame = pd.DataFrame(datos)
frame
```

```python
#La función head selecciona solo las primeras cinco filas

frame.head()
```

```python
#La función tail selecciona las últimas cinco

frame.tail()
```

```python
#Se pueden especificar el orden en como aparezcan las columnas en nuestro DataFrame

pd.DataFrame(datos, columns=["año", "estado", "población (millones)"])
```

```python
#Para renombrar las columnas utilizando la función rename

frame.rename(columns={"estado": "Estado",
                      "año": "Año",
                      "población (millones)": "Población (Millones)"},
             inplace=True) #Inplace modifica el dataframe directamente en vez de crear uno nuevo
frame
```

```python
#Una columna en un DataFrame se puede recuperar como una serie

frame["Estado"] #Se pueden recuperar usando notación de lista: frame["nombre de la columna"]
```

```python
frame.Año #O también usando la notación de punto: frame.nombredelacolumna
```

```python
#También se pueden ordenar según como se requiera

frame = frame.sort_values(by="Población (Millones)", ascending=False) #Ordenamos el DataFrame de mayor a menor según su población
frame
```
