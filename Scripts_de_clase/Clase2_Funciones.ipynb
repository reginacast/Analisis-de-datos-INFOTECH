```python
#Función simple
def bienvenida():
  print("Bienvenidos al sistema")

bienvenida()
```

```python
#Función con argumentos
def saludar(nombre):
  print("Hola", nombre)

saludar("Pablo")
```

```python
#Función con argumentos por posición
def resta(a, b):
  return a - b

resultado1 = resta(10, 3)
print("El resultado de la resta por posición es", resultado1)

#Función con argumentos por nombre
resultado2 = resta(b=15, a=12)
print("El resultado de la resta por nombre es", resultado2)
```

```python
#Función con argumentos por defecto
def calcular_total(precio, impuesto=0.16):
  return precio + (precio * impuesto)

print("Total con impuesto por defecto: ", calcular_total(100))
print("Total con impuesto personalizado: ", calcular_total(100, 0.10))
```

```python
#Función con argumentos de longitud variable
def sumar_numeros(*numeros): # El asterisco (*) indica que 'numeros' capturará todos los argumentos posicionales en una tupla. De lo contrario, al escribir solo 'numeros' Python piensa que solo le daremos un argumento y no una tupla.
  total = 0
  for n in numeros: # Itera sobre cada número en la tupla 'numeros'
    total += n
  return total

# Cuando llamas a la función con múltiples valores, estos se agrupan en la tupla 'numeros'
print("Suma de varios numeros: ", sumar_numeros(1,2,3,4,5))
```

```python
#Función con la sentencia return
def mensaje_final():
  return "Gracias, el programa ha finalizado correctamente"

print(mensaje_final())
```
