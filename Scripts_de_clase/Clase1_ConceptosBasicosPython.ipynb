ðŸ§© bucle while

```python
x = 5
while x > 0:
    x -= 1
    print(x)
else:
    print("Fin del ciclo")
```

```python
from collections.abc import Iterable #Se llama esta biblioteca
lista = [1, 2, 3]
cadena = "Python"
numero = 10

print(isinstance(lista, Iterable)) #Funcion usada para determinar si el dato es iterable o no
print(isinstance(cadena, Iterable))
print(isinstance(numero, Iterable))
```

ðŸ§© for anidado

```python
lista = [
    [34, 25, 12],
    [12, 65, 34],
    [90, 71, 23]
]

for fila in lista:
    for elemento in fila:
        print(elemento)
```

ðŸ§© while anidado

```python
i = 0
j = 0
while i < 3:
    while j < 3:
        print(i, j)
        j += 1
    i += 1
    j = 0
```
