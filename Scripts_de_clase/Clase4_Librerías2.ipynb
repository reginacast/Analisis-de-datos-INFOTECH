## **Matplotlib:**

Permite generar gráficos con pocas líneas de código.

- **Gráfico:** es un dibujo bidimensional o tridimensional que muestra una relación mediante puntos, una curva, una serie de barras, etc.

- **Ejes:** El eje X (horizontal) representa los valores independientes y el eje Y (vertical) corresponde a los valores dependientes.

Esta librería se organiza jerárquicamente. Aprecen primero las figuras y luego los ejes:

```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Graficar datos
plt.plot([-1, -4.5, 16, 23]) #le damos valores en el eje Y
plt.show()
```

```python
#Podemos cambiar el tipo de gráfico que queremos mostrar
plt.plot([-1, -4.5, 16, 23], "ob") #'o' representa que sea un gráfico de puntos y 'b' el color azul
plt.show()
```

```python
#Creamos un gráfico más elaborado

dias = range(1, 9) #toma un rango de valores entre 1 y 9 (excluyendo este último valor)
valores_celsius = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots() # Se crea una figura con un eje.
ax.plot(dias, valores_celsius) #Le damos los datos a tomar
ax.set(xlabel='Día', #Agregamos etiquetas
       ylabel='Temperatura en Celsius',
       title='Gráfico de Temperatura')

plt.show() #Mostramos la gráfica
#fig.savefig("Figura 1.png") #Guardamos la imagen de la gráfica si lo deseamos
```

```python
#Podemos verificar y definir los rangos de los ejes

dias = range(1, 9)
valores_celsius = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots() # Se crea una figura con un eje.
ax.plot(dias, valores_celsius)
ax.set(xlabel='Día',
       ylabel='Temperatura en Celsius',
       title='Gráfico de Temperatura')

print("Los límites actuales para los ejes son:")
print(ax.axis()) #Se imprimen los valores de los ejes actuales

print("Se establecen nuevos ejes con los siguientes valores:")
xmin, xmax, ymin, ymax = 0, 10, 14, 45 #Se guardan los valores en variables
print(xmin, xmax, ymin, ymax)

ax.axis([xmin, xmax, ymin, ymax]) #Se ponen los nuevos valores ahora en la gráfica
plt.show()
```

```python
#Podemos representar gráficamente varias figuras a la vez

dias = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

fig, ax = plt.subplots()
ax.plot(dias, celsius_min, color="blue") #Modificamos el color de la línea
ax.plot(dias, celsius_min, "ob") #Le decimos que queremos puntos de color azul
ax.plot(dias, celsius_max, color="red")
ax.plot(dias, celsius_max, "or")

ax.set(xlabel='Día', #Agregamos el título y las etiquetas en los ejes
       ylabel='Temperatura en Celsius',
       title='Gráfico de Temperatura')

plt.show()
```

### **Tipos de gráficos**

**1. Gráfico de barras:** Se utiliza para datos categóricos (categorías con nombre). Compara grupos que no tienen una relación numérica necesaria entre sí

```python
años = [str(año) for año in range(2010, 2021)] #Iteramos para eficientar el proceso y obtener los años rápidamente
visitantes = (
1241, 50927, 162242,
222093, 665004, 2071987,
2460407, 3799215, 5399000,
5474016, 6003672)

plt.bar(años, visitantes, color="green")
plt.xlabel("Años")
plt.ylabel("Valores")
plt.title("Ejemplo de Gráfica de Barras")
plt.plot()

plt.show()
```

**2. Histogramas:** Se utiliza para datos cuantitativos continuos (números que fluyen en una escala). Muestra la distribución de una sola variable numérica.

```python
datos = np.random.normal(size=10000) #Genera un arreglo unidimensional de 10000 números aleatorios
plt.hist(datos, bins=20) #Lo dividimos en 20 barras
plt.title("Ejemplo de Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")

plt.show()
```

**3. Diagrama de dispersión:** Su objetivo principal es mostrar la relación o correlación entre dos variables numéricas.

```python
x = np.arange(0, 11) # Crea una lista de números del 0 al 10 para el eje X

# Crea 3 series de datos aleatorios para el eje Y:
y1 = np.random.randint(2, 7, (11,)) # Genera 11 números enteros entre el 2 y el 6
y2 = np.random.randint(9, 14, (11,)) # Genera 11 números enteros entre el 9 y el 13
y3 = np.random.randint(15, 25, (11,)) # Genera 11 números enteros entre el 15 y el 24

plt.scatter(x, y1, label="y1")#Dibuja los puntos de y1 (usa el color azul y punto circular por defecto)
plt.scatter(x, y2, marker='v', color='r', label="y2") #Dibuja y2 usando un triángulo hacia abajo ('v') y color rojo ('r')
plt.scatter(x, y3, marker='^', color='m', label="y3") #Dibuja y3 usando un triángulo hacia arriba ('^') y color magenta ('m')

plt.legend()#Muestra la leyenda (el cuadro que indica qué es y1, y2 e y3)
plt.title('Ejemplo de Diagrama de Dispersión')

plt.show()
```

**4. Gráfico apilado:** Su objetivo es mostrar la composición de un total. Se usa para ver cómo diferentes partes forman un todo y cómo ese total cambia.

```python
x = [ 1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 44, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]

plt.stackplot(x, y1, y2, y3)
plt.title('Ejemplo de Gráfico Apilado')
plt.show()
```

**5. Gráfico de pastel:** Es una representación circular diseñada para mostrar proporciones y porcentajes. Este no tiene ejes tradicionales; en su lugar, el círculo completo representa el 100% de un total, y cada "rebanada" representa una parte de ese todo.

```python
# Gráfico circular, donde las porciones se ordenan y trazan en sentido antihorario

# Datos de ejemplo
etiquetas = 'C', 'Python', 'Java', 'C++', 'C#' # Define los nombres de las categorías
tamaños = [13.38, 11.87, 11.74, 7.81, 4.41]  # Define los valores (porcentajes o cantidades)
explotar = (0, 0.1, 0, 0, 0)  # Define el desfase: 0.1 separa la 2ª rebanada (Python)

fig1, ax1 = plt.subplots() # Crea la figura y el eje donde se dibujará

# Generación del gráfico de pastel con sus parámetros principales:
ax1.pie(tamaños,               # Datos numéricos de las porciones
        explode= explotar,     # Aplica el efecto de separación a las porciones
        labels= etiquetas,     # Asigna los nombres a cada rebanada
        autopct='%1.1f%%',     # Formatea y muestra el porcentaje con un decimal
        startangle=0)          # Rota el inicio del gráfico (0 grados es a la derecha)

ax1.axis('equal') # Forza a que el círculo sea redondo y no un óvalo
plt.title('Índice TIOBE para Mayo 2021')
plt.show()
```
